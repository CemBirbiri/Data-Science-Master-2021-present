We have 2 datasets so we need to build two different linear regression model.

- The second dataset is: "regm.txt" and we will cal it DATASET-2.

#DATASET-1 and MODEL-1

Model-1 is the linear regression model that I will train on dataset-1.

Load Dataset-1 and name its columns:

```python
import pandas as pd

colnames_1=['feature_1','target']
dataset1 = pd.read_csv('/content/reg1(1).txt', names=colnames_1, header=None, sep=' ')
```

See the dataset1:

```python
dataset1.head(3)
```

See the shape:

```python
dataset1.shape
```

Let's define X (feature vectors) and y (target vector)

```python
#X will be our training and validation set
X = dataset1.drop(['target'], axis=1)

#y will be the corresponding class labels
y = dataset1['target']

X.head(3)
```

```python
y.head(3)
```

We will make 10 experiments to observe the Linear Regression model performance. In each experiment we split the data as validation and train with 30% - 70% portions, respectively. While splitting we shuffle the dataset each time. Also, the "random_state" parameter is set to 31 to have eliminate randomness in each run.

After, we define a Linear Regression model. Then we fit the model with training data and do the prediction on validation data.

At the end we measure the model performance. Accuracy (e.g. classification accuracy) is a measure for classification, not regression so we can't calculate accuracy for a regression model. For regression, we will use three metrics on both train and validaton datasets and calculate train and test error:

1. Mean Absolute Error (mae)
2. Mean Squared Error (mse)
3. Root Mean Squared Error (rmse)

(Instead of using sklearn Cross-validation module, I wrote the cross-validation code in a for-loop with # of iterations is 10.)

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np
from sklearn.model_selection import train_test_split



for i in range(1,11):
  #Train-val split
  X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.30, shuffle=True, random_state=31)
  
  #Our linear regression model
  regressor = LinearRegression()

  #Fit and predict
  r= regressor.fit(X_train, y_train)
  y_pred = regressor.predict(X_val)
  y_pred_train = regressor.predict(X_train)

  #Print the errors:
  print("Experiment -",i)
  print('VAL mae:', metrics.mean_absolute_error(y_val, y_pred))
  print('TRAIN mae:', metrics.mean_absolute_error(y_train, y_pred_train))
  print( "Train > Val" if metrics.mean_absolute_error(y_train, y_pred_train)>metrics.mean_absolute_error(y_val, y_pred) 
        else  "Val > Train")
  print()
  print('VAL mse:', metrics.mean_squared_error(y_val, y_pred))
  print('TRAIN mse:', metrics.mean_squared_error(y_train, y_pred_train))
  print( "Train > Val" if metrics.mean_squared_error(y_train, y_pred_train) > metrics.mean_squared_error(y_val, y_pred)
        else "Val > Train")
  print()
  print('VAL rmse:', np.sqrt(metrics.mean_squared_error(y_val, y_pred)))
  print('TRAIN rmse:', np.sqrt(metrics.mean_squared_error(y_train, y_pred_train)))
  print( "Train > Val" if np.sqrt(metrics.mean_squared_error(y_train, y_pred_train)) > np.sqrt(metrics.mean_squared_error(y_val, y_pred))
        else "Val > Train")
  print()
  print("----------------------------------------------------------------")
  
```

###Which error is better and why?

In Dataset-1 for all of the error types, the training error is greater than validation error. This can be explained as the linear regression model is unable to learn from dataset. There can be two reasons:
1. The training set had many 'hard' cases to learn.
2. The validation set had mostly 'easy' cases to predict.
